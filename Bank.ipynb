## combining values from different regions

```python
'''
AP = Asia Pacific
CEE = Central and Eastern Europe
LA = Latin America 
MEA = Middle East and Africa
NA = North America
WE = Western Europe
'''
```

```python
asia_pacific_df = pd.read_csv('../machine_learning_project/Data cleaned/AP_value.csv', index_col='Country')
```

```python
def get_cols(df):
    return df.columns

for col in get_cols(asia_pacific_df):
    asia_pacific_df[col] = asia_pacific_df[col].str.replace('$', '')

asia_pacific_df['region'] = 'Asia Pacific'

asia_pacific_df
```
